```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
def my_sigmoid(X):
    try: 
        f = 1 / (1+np.exp(-X))
    except Exception as e:
        print(e)
    else:
       return f
```

```python
def my_tanh(X):
  try: 
      f = (np.exp(X)-np.exp(-X))/(np.exp(X)+np.exp(-X))
  except Exception as e:
      print(e)
  else:
      return f
```

```python
def my_softsign(X):
  try: 
      f = X/(np.abs(X)+1)
  except Exception as e:
      print(e)
  else:
      return f
```

```python
def my_relu(X):
  if X >= 0:
    return X
  else:
    return 0
```

```python
def my_softplus(X):
  try: 
      f = np.log(np.exp(X)+1)
  except Exception as e:
      print(e)
  else:
      return f  
```

```python
# Plot Sigmoid Function
X = np.arange(-10,10,0.1)
y = my_sigmoid(X)
plt.plot(X,y)
plt.grid()
```

```python
# Plot Sigmoid Function
X = np.arange(-10,10,0.1)
y = my_sigmoid(X)
y_div = y*(1-y)
X_norm = X/10
plt.plot(X,y)
plt.plot(X,y_div)
# plt.plot(X, X_norm)
plt.grid()
```

```python
# plot tanh function
X = np.arange(-10,10,0.1)
y = my_tanh(X)
plt.plot(X,y)
plt.grid()
```

```python
# plot tanh function
X = np.arange(-10,10,0.1)
y = my_tanh(X)
y_div = 1 - y**2
plt.plot(X,y)
plt.plot(X,y_div)
plt.grid()
```

```python
# plot softsign function
X = np.arange(-10,10,0.1)
y = my_softsign(X)
y_div = 1 / (np.abs(X)+1)**2
plt.plot(X,y)
plt.plot(X,y_div)
plt.grid()
```

```python
# plot relu function
X = np.arange(-3,3,0.1)
y = [my_relu(x) for x in X]
y_div = [1 if x >= 0 else 0 for x in X ]
plt.plot(X,y)
plt.plot(X,y_div)
plt.grid()
```

```python
# plot softplus function
X = np.arange(-3,3,0.1)
y = my_softplus(X)
y_div = 1 / (np.exp(-X)+1)
plt.plot(X,y)
plt.plot(X,y_div)
plt.grid()
```
